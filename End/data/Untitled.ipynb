```python
from collections import defaultdict
import json

f=open("routes.txt")
routes=defaultdict(list)

for line in f.readlines():
    start, destination, length, color = line.strip().split(" ")
    routes[start].append((destination, length, color))

f.close()
```

```python
routes
```

```python
data = {}
data["routes"]= []

for k,v in routes.items():
    city = {}
    city["name"] = ["String", k]
    city["neighbours"] = []
    cityRoutes = []
    for route in v:
        cityRoute = {}
        cityRoute["city"] = ["String", route[0]]
        cityRoute["routeColor"] = ["String", route[2]]
        cityRoute["routeLength"] = ["Integer", int(route[1])]
        cityRoutes.append(cityRoute)
    city["neighbours"] = cityRoutes
    data["routes"].append(city)
```

```python
with open('routes.json', 'w') as outfile:
    json.dump(data, outfile)
```

# Routes


```python
f=open("routeCards.txt")
routeCards = []
for line in f.readlines():
    routeCards.append(line.strip().split("\t"))

f.close()
```

```python
routeCards[0]
```

```python
cardData = {}
cardData["cards"] = []

for routeCard in routeCards:
    card = {}
    card["start"] = ["String", routeCard[0].replace(" ", "")]
    card["destination"] = ["String", routeCard[1].replace(" ", "")]
    card["points"] = ["Integer", int(routeCard[2])]
    cardData["cards"].append(card)
```

```python
cardData
```

```python
with open('cards.json', 'w') as outfile:
    json.dump(cardData, outfile)
```
